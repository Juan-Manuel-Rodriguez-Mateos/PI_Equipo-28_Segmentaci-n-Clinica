```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = '/content/drive/MyDrive/Proyecto integrador/Analísis EDA/WAHA_Urolithins_MetabotypesData.dta'
df = pd.read_stata(ruta)
```

```python
# Vista general de los datos:

print("🔹 Dimensiones del DataFrame:", df.shape)
print("\n🔹 Primeras filas:")
display(df.head())
```

```python
#Tipo de datos:

# Paso 5: Tipos de datos
print("\n🔹 Tipos de datos:")
print(df.dtypes)
```

```python
# Valores nulos:

print("\n🔹 Valores nulos por columna:")
print(df.isnull().sum())
```

```python
# Estadísticas descriptivas
print("\n🔹 Estadísticas descriptivas:")
display(df.describe())
```

```python
# Visualización de distribución de algunas variables numéricas
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for col in num_cols[:4]:  # Puedes ajustar el número de columnas a graficar
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()
```

```python
# Detección de valores atípicos con boxplots
for col in num_cols[:4]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot de {col}")
    plt.tight_layout()
    plt.show()
```

```python
# Análisis bivariado (correlación)

uro_vars = [col for col in df.columns if 'uro' in col.lower()]
delta_vars = [col for col in df.columns if 'delta' in col.lower() or 'change' in col.lower()]

print("Variables de urolitinas encontradas:", uro_vars)
print("Variables de deltas cognitivas encontradas:", delta_vars)

if uro_vars and delta_vars:
    corr = df[uro_vars + delta_vars].corr().loc[uro_vars, delta_vars]

    plt.figure(figsize=(1.2 * len(delta_vars), 0.6 * len(uro_vars)))

    sns.heatmap(
        corr,
        annot=True,
        cmap="coolwarm",
        fmt=".2f",
        annot_kws={"size": 7},
        cbar_kws={"shrink": 0.8}
    )

    plt.title("Correlación Urolitinas vs Deltas Cognitivos")
    plt.xticks(rotation=45, ha="right", fontsize=8)
    plt.yticks(rotation=0, fontsize=8)
    plt.tight_layout()
    plt.show()

else:
    print("No se encontraron variables suficientes para correlacionar.")
```

```python
# Análisis de una variable categórica si existe
cat_cols = df.select_dtypes(include=['category', 'object']).columns.tolist()
for col in cat_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribución de {col}")
    plt.tight_layout()
    plt.show()
```
